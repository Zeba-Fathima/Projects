## Trigger Function

**AUTO DUE DATE GENERATOR** 

on the entry of book in issue table given the issue date the due date for the same is autofilled, no user input required

```python
cur.execute("""
CREATE OR REPLACE FUNCTION calculate_due_date()
RETURNS TRIGGER AS $$
BEGIN
    -- Set Due_Date as 14 days ahead of Issue_Date
    NEW.Due_Date = NEW.Issue_Date + INTERVAL '14 days';
    RETURN NEW;
END;
$$ LANGUAGE plpgsql;

CREATE TRIGGER set_due_date
BEFORE INSERT ON Book_Issue
FOR EACH ROW
EXECUTE FUNCTION calculate_due_date();
""")
con.commit()
```

**AUTO FINE CALCULATOR**

when a certain book with a certain issue_id is updated to be retured the fine is calculated automatically, for each day past the due day the fine amount increases by Rs.5 with the comparison of due date and the entered return date

```python
sql = """
CREATE OR REPLACE FUNCTION calculate_fines()
RETURNS TRIGGER AS $$
BEGIN
    IF NEW.Is_Return = TRUE THEN
        IF NEW.Return_Date::DATE < NEW.Due_Date THEN
            UPDATE Book_Issue
            SET Return_Date = NEW.Return_Date::DATE, Fine_Amount = 0
            WHERE Issue_Id = NEW.Issue_id;
        ELSE
            UPDATE Book_Issue
            SET Return_Date = NEW.Return_Date::DATE, Fine_Amount = (NEW.Return_Date::DATE - NEW.Due_Date) * 5
            WHERE Issue_Id = NEW.Issue_id;
        END IF;
    END IF;
    RETURN NEW;
END;
$$ LANGUAGE plpgsql;

CREATE TRIGGER auto_calculate_fines
AFTER UPDATE OF Is_Return ON Book_Issue
FOR EACH ROW
EXECUTE FUNCTION calculate_fines();
"""
cur.execute(sql)
con.commit()
```

**CHECK - BORROWING LIMIT**

limits the insert of books for the user with more that 2 books under their id

```python
sql = """
CREATE OR REPLACE FUNCTION check_borrowing_limit()
RETURNS TRIGGER AS $$
DECLARE
    borrowed_count INT;
BEGIN
    -- Count the number of books currently borrowed and not returned by the user
    SELECT COUNT(*) INTO borrowed_count
    FROM Book_Issue
    WHERE User_id = NEW.User_id AND Is_Return = FALSE;

    IF borrowed_count >= 3 THEN
        -- Raise an error message
        RAISE EXCEPTION 'This member (%s) has reached the borrowing limit. Further borrowing is restricted.', NEW.User_id;
        RETURN NULL; -- Exiting the function
    END IF;

    RETURN NEW;
END;
$$ LANGUAGE plpgsql;

-- Create the trigger to execute the check_borrowing_limit function after each INSERT
CREATE TRIGGER enforce_borrowing_limit
AFTER INSERT ON Book_Issue
FOR EACH ROW
EXECUTE FUNCTION check_borrowing_limit();
"""
cur.execute(sql)
con.commit()
```

**UPDATE Avaialility on Issue**

aims to automate the update the availability_status of the book once it has been issued

```python
sql = """CREATE OR REPLACE FUNCTION update_availability_on_issue()
RETURNS TRIGGER AS $$
BEGIN
    -- Update the Availability_Status to 'Not Available' when a book is issued
    UPDATE Book
    SET Availability_Status = 'Checked Out'
    WHERE Book_id = NEW.Book_id;

    RETURN NEW;
END;
$$ LANGUAGE plpgsql;

CREATE TRIGGER set_availability_on_issue
AFTER INSERT ON Book_Issue
FOR EACH ROW
EXECUTE FUNCTION update_availability_on_issue();"""

cur.execute(sql)
con.commit()
```

**UPDATE Availability on Return**

aims to automate the update the availability_status of the book once it has been returned.

```python
sql = """
CREATE OR REPLACE FUNCTION update_availability_on_return()
RETURNS TRIGGER AS $$
BEGIN
    IF NEW.Is_Return = TRUE THEN
        UPDATE Book
        SET Availability_Status = 'Available'
        WHERE Book_id = NEW.Book_id;
    END IF;

    RETURN NEW;
END;
$$ LANGUAGE plpgsql;

CREATE TRIGGER book_return_trigger
AFTER UPDATE OF Is_Return ON Book_Issue
FOR EACH ROW
EXECUTE FUNCTION update_availability_on_return();
"""
cur.execute(sql)
con.commit()
```

## Views

**Books To Return**

This aims to give the view of all the remaining books that needs to be returned to the library that has been issued to a particular.

```python
sql = """CREATE OR REPLACE VIEW books_to_return AS
                SELECT
                    bi.book_id,
                    bi.due_date,
                    bi.user_id
                FROM
                    book_issue bi
                WHERE
                    bi.is_return = false;"""
cur.execute(sql)
con.commit()
```

**Checked Out Books**

aims to give the all the checked out books in the library

```python
sql = """
CREATE MATERIALIZED VIEW checked_out_books AS
SELECT bi.Issue_id, bi.Book_id, b.Title, u.Name AS User_Name, bi.Issue_Date, bi.Due_Date, bi.Is_Return, bi.Return_date, bi.Fine_Amount
FROM Book_Issue bi
JOIN Book b ON bi.Book_id = b.Book_id
JOIN Users u ON bi.User_id = u.User_id
WHERE bi.Is_Return = FALSE;
"""
cur.execute(sql)
con.commit()
```

**Available Books**

aimd to give/show all the books that are currently available in the library.

```python
view_sql = """
CREATE OR REPLACE VIEW available_book AS
SELECT *
FROM Book
WHERE Availability_Status = 'Available';
"""
cur.execute(view_sql)
con.commit()
```

**Users Summary**

this aims to give the varied users present in the library with their total count.
as of now it is just member and librarian

```python
sql= """
CREATE OR REPLACE VIEW user_roles_summary AS
SELECT Role, COUNT(*) AS  UserCount
FROM Users
GROUP BY Role;
"""
cur.execute(sql)
con.commit()
```

# User Interface

* It is a command line based interface
* Actors: Member and Librarian
* Both authenticated on entering to the system
* on successful authentication both actors proceed to performing their actions
* Actions:
    1. Member Actions:
    
         * View of Over Due Books
         * Search Books Based On Title, Subject, Author --> ILIKE operations
         * View of Available Books
         
    2. Librarian Actions: 
    
        i. Book Management:
            * Add Book --> New Book record is added : INSERT
            * Update Book --> edit the existing row info : UPDATE
            * Remove Book --> Delete a record from table: DELETE
            
       ii. User Management
            * Add User --> New User record is added : INSERT
            * Update User --> edit the current existing user info : UPDATE
            * Remove User --> Delete a user from the table : DELETE
            
      iii. Issue and Returns: 
            * Issue --> Insert record into issue table, invoking triggers
            * Return --> updates status, invoking triggers
            
       iv. Search: Search record based on title, author, subject
       
        v. View the checkout books
        
       vi. View User Role summary
       
       
* on completing their respective operation as required, they can exit the program : Logout
